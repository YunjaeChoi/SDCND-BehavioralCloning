```python
import numpy as np
import pandas as pd

#reading data, data1:first track data, data2: second track data
data1 = pd.read_csv('data/driving_log.csv')
data2 = pd.read_csv('data2/driving_log.csv')
data = pd.concat([data1,data2],ignore_index=True)
data
```

```python
#maching file locations
center_file_loc = pd.Series(data['center_image'])
center_file_loc = center_file_loc.str.replace('/home/yjchoi/Desktop/Udacity-Selfdriving Car ND','/home/lics/mycode')

left_file_loc = pd.Series(data['left_image'])
left_file_loc = left_file_loc.str.replace('/home/yjchoi/Desktop/Udacity-Selfdriving Car ND','/home/lics/mycode')

right_file_loc = pd.Series(data['right_image'])
right_file_loc = right_file_loc.str.replace('/home/yjchoi/Desktop/Udacity-Selfdriving Car ND','/home/lics/mycode')
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline


imgs = []
angle = []

for i in range(len(data)):
    #all images are augmented by flipping horizontally
    #center img
    img = mpimg.imread(center_file_loc.iloc[i])
    img_flipped = np.fliplr(img)
    imgs.append(img)
    imgs.append(img_flipped)
    
    angle.append(data['angle'].iloc[i])
    angle.append(-data['angle'].iloc[i])
    
    #angle_adjust for left and right images, for sharp turns angle multipling factor was used.
    if np.abs(data['angle'].iloc[i]) < 0.5:
        angle_adjust = 0.2
        angle_mul_fac = 1.0
    else:
        angle_adjust = 0.2
        angle_mul_fac = 1.15
    
    #left img
    img = mpimg.imread(left_file_loc.iloc[i])
    img_flipped = np.fliplr(img)
    imgs.append(img)
    imgs.append(img_flipped)
    
    left_angle = min((data['angle'].iloc[i] * angle_mul_fac) + angle_adjust,1.)
    angle.append(left_angle)
    angle.append(-left_angle)
    
    #right img
    img = mpimg.imread(right_file_loc.iloc[i])
    img_flipped = np.fliplr(img)
    imgs.append(img)
    imgs.append(img_flipped)
    
    right_angle = max((data['angle'].iloc[i] * angle_mul_fac) - angle_adjust,-1.)
    angle.append(right_angle)
    angle.append(-right_angle)

imgs = np.array(imgs)
angle = np.array(angle)

print(imgs.shape,angle.shape)
```

```python
#checking shapes
X_train = imgs
Y_train = angle[:,None]
print(X_train.shape,Y_train.shape)
```

```python
from keras.layers import Conv2D, MaxPooling2D, Cropping2D
from keras.layers import Dropout, Flatten, Dense, Input, Lambda, Activation, BatchNormalization
from keras import regularizers
from keras.models import Model

input_shape = imgs.shape[1:]
nb_outputs = 1 #angle

# 5 conv layers followed by max pooling and dropout, additional 1 dense layer for the output
inputs = Input(shape=input_shape)
h = Lambda(lambda x: (x * 5. / 255.) - 2.5)(inputs)
h = Cropping2D(cropping=((50,25), (0,0)))(h)
h = Conv2D(8,5,padding='same',activation='relu')(h)
h = MaxPooling2D(pool_size=2)(h)
h = Dropout(0.2)(h)
h = Conv2D(16,5,padding='same',activation='relu')(h)
h = MaxPooling2D(pool_size=2)(h)
h = Dropout(0.2)(h)
h = Conv2D(32,3,padding='same',activation='relu')(h)
h = MaxPooling2D(pool_size=2)(h)
h = Dropout(0.2)(h)
h = Conv2D(64,3,padding='same',activation='relu')(h)
h = MaxPooling2D(pool_size=2)(h)
h = Dropout(0.2)(h)
h = Conv2D(128,3,padding='same',activation='relu')(h)
h = MaxPooling2D(pool_size=2)(h)
h = Flatten()(h)
outputs = Dense(nb_outputs,activation='tanh')(h)
model = Model(inputs,outputs)
model.summary()
model.compile(optimizer='adam', loss='mse')
```

```python
from keras.callbacks import ReduceLROnPlateau
from keras.callbacks import ModelCheckpoint

#save model
filepath="saved_model/weights.best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='auto')
reduce_lr = ReduceLROnPlateau(monitor = 'val_loss',factor = 0.2,patience = 3,verbose=1,min_lr = 0.00005)
#train
epochs = 10
history_object = model.fit(X_train, Y_train, validation_split=0.2, shuffle=True, 
                           epochs=epochs, batch_size=32,callbacks=[checkpoint,reduce_lr])
```

```python
print(history_object.history.keys())
```

```python
#plot loss
plt.plot(history_object.history['loss'])
plt.plot(history_object.history['val_loss'])
plt.title('model mean squared error loss')
plt.ylabel('mean squared error loss')
plt.xlabel('epoch')
plt.legend(['training set', 'validation set'], loc='upper right')
plt.show()
```
